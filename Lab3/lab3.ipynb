```python
# Create a Car class with attributes make , model , and year . Add a method to display car
# details.
class Car:
    def __init__(self,make,model,year):
        self.make=make
        self.model=model
        self.year=year
    def __str__(self):
        return f"Manufacture by : {self.make.upper()}\nModel : {self.model}\nYear Made : {self.year}"
C1=Car("Tesla","Model A",2023)
print(C1)
```

```python
# Extend the Car class to ElectricCar with an additional attribute battery_capacity . Add a
# method to display battery info.
class ElectricCar(Car):
    def __init__(self, make, model, year, battery_capacity):
        super().__init__(make, model, year)
        self.battery_capacity = battery_capacity
    
    def display_battery_info(self):  # A dedicated method as requested
        return f"Battery Capacity: {self.battery_capacity} kWh"
    
    def __str__(self):
        return f"{super().__str__()}\n{self.display_battery_info()}"

EC1 = ElectricCar("Tesla", "Model 3", 2024, 79)
print(EC1)  
EC1.display_battery_info()
```

```python
class Rectangle:
    def __init__(self,length,breadth):
        self.length=length
        self.breadth=breadth
    def __str__(self):
        return f"Rectangle({self.length}cm,{self.breadth}cm)"
    def CalArea(self):
        return f"Area of the {str(self)} ={self.length*self.breadth} sq.cm"
    def CalPreme(self):
        return f"Premeter of the {str(self)}= {2*(self.length+self.breadth)} cm"
R=Rectangle(5,4)
print(R)
print(R.CalArea())
print(R.CalPreme())
```
