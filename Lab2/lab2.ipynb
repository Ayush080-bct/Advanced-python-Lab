```python
class CustomCounter:
    def __init__(self, limit):
        self.limit=limit
        self.current=0
    def __iter__(self):
        return self
    def __next__(self):
        if self.current<self.limit:
            self.current+=1
            return self.current
        else:
            raise StopIteration
counter=CustomCounter(5)
for num in counter:
    print(num,end=' ')
```

```python
# Write an iterator that returns the first n even numbers.
class   EvenCounter:
    def __init__(self, limit):
        self.limit=limit
        self.current=0
        self.count=1
    def __iter__(self):
        return self
    def __next__(self):
        if self.count<=self.limit:
            
            self.count+=1
            m=self.current
            self.current+=2
            return m
        else:
            raise StopIteration
counter=EvenCounter(5)
for num in counter:
    print(num,end=' ')
```

```python
# Write a generator that yields the Fibonacci sequence up to n terms.
def fibb(n):
    a,b=0,1
    for _ in range(n):
        yield a
        a,b=b,a+b
list(fibb(10))
```

```python
# Implement an iterator class that iterates over a range in reverse order.
class ReverseOrder:
    def __init__(self,limit):
        self.limit=limit
        self.current=limit
    def __iter__(self):
        return self
    def __next__(self):
        if self.current >0:
            self.current-=1
            return self.current
        else:
            raise StopIteration
R_order=ReverseOrder(5)
for i in  R_order:
    print(i)
```

```python
# Create a decorator to measure memory usage of a function.

import tracemalloc
from functools import wraps#it has it own importance , let understand first with this function
#  this helps to keep original function identity
def measure_memory(func):
    @wraps(func)#it contains meta data,name,doc etc
    def wrapper(*args, **kwargs):
        tracemalloc.start()
        
        result = func(*args, **kwargs)
        
        current, peak = tracemalloc.get_traced_memory()
        tracemalloc.stop()
        
        print(f"Memory usage of {func.__name__}:")
        print(f"  Current: {current / 10**6:.4f} MB")
        print(f"  Peak: {peak / 10**6:.4f} MB")
        
        
    return wrapper
@measure_memory
def test():
        return [i for i in range(1_000_000)]
test()
print(test.__name__)
# output of print(test.__name__) is test which is correct with the help of func tools
```

```python
# Write a generator that yields prime numbers indefinitely.
def Prime_numbers():
    n=2
    while True:
        is_prime=True
        for i in range(2,n//2+1):
            if(n%i==0):
                is_prime=False
                break
        if is_prime:
            yield n
        n+=1
gen=Prime_numbers()
print(type(gen))
prime=[next(gen) for _ in range(50)]#first 50 number(prime)
print(prime)
```

### cache
- store somthing temporarily so you dont have to cache result of a function in a simple terms.
- In programming , caching a fucntions's result means :
 - The first time you call the function , it runs normally and produces a result.
 - That result is saved(cached).
 - The next time you calll the function with same input , instead of running all code again , it just returns the saved result instantly.
**Why cache is useful**
- Saves time avoid repeating expensive calculation (like generating primes,fetching data,doing heaavy math  )
- save resourses :less cpu ,memory , or network usage.
- Makes program faster when the same inputs are used multiple times.

```python
#Example with out cache
def slow_square(n):
    print("calculating")
    return n*n
print(slow_square(5))#run once
print(slow_square(5))#recalculate everything 
```

```python
cache={}
def slow_square(n):
    if n in cache:
        return cache[n]
    print("Calculating....")
    result=n*n
    cache[n]=result
    return result
print(slow_square(5))
print(slow_square(5))## returns cached result 
print(slow_square(6))
print(slow_square(6))
```

**Above output shows** 
- if it is already calculated then it just return the value not recalculate everything

### Memonization 
*which we did in calcualting Fibb series , we did save result in cache so not to calculate everything for same input*
- It’s a technique where you remember (cache) the results of function calls.
- If the function is called again with the same input, you don’t recompute — you just return the stored result.
- This makes recursive algorithms (like Fibonacci) much faster because they avoid repeating the same work.


```python
table = {}
a = []

def fibM(n):
    if n == 0 or n == 1:
        return n
    else:
        if n not in table:              # check cache
            table[n] = fibM(n-1) + fibM(n-2)  # compute and store in cache
        return table[n]

n = int(input("enter the number of fib series: "))
for i in range(n):
    a.append(fibM(i))

print(*a)
print(table)
```

```python
# Write an iterator that yields only unique items from a list.
Alists=[1,2,1,4,2,1,5,4,3,5]

def Uniq_items(Alist):
    seen=set()
    for Alist in Alists:
        if Alist not in seen:
            seen.add(Alist)
            yield Alist
new_alist=Uniq_items(Alists)
print(next(new_alist))
print(next(new_alist))
print(next(new_alist))
unique_items_from_list = list(new_alist)
print(sorted(unique_items_from_list))
```
