````python
class CustomCounter:
    def __init__(self, limit):
        self.limit=limit
        self.current=0
    def __iter__(self):
        return self
    def __next__(self):
        if self.current<self.limit:
            self.current+=1
            return self.current
        else:
            raise StopIteration
counter=CustomCounter(5)
for num in counter:
    print(num,end=' ')
````

````python
# Create a decorator to measure memory usage of a function.

import tracemalloc
from functools import wraps#it has it own importance , let understand first with this function
#  this helps to keep original function identity
def measure_memory(func):
    @wraps(func)#it contains meta data,name,doc etc
    def wrapper(*args, **kwargs):
        tracemalloc.start()
        
        result = func(*args, **kwargs)
        
        current, peak = tracemalloc.get_traced_memory()
        tracemalloc.stop()
        
        print(f"Memory usage of {func.__name__}:")
        print(f"  Current: {current / 10**6:.4f} MB")
        print(f"  Peak: {peak / 10**6:.4f} MB")
        
        
    return wrapper
@measure_memory
def test():
        return [i for i in range(1_000_000)]
test()
print(test.__name__)
# output of print(test.__name__) is test which is correct with the help of func tools
````

### cache
- store somthing temporarily so you dont have to cache result of a function in a simple terms.
- In programming , caching a fucntions's result means :
 - The first time you call the function , it runs normally and produces a result.
 - That result is saved(cached).
 - The next time you calll the function with same input , instead of running all code again , it just returns the saved result instantly.
**Why cache is useful**
- Saves time avoid repeating expensive calculation (like generating primes,fetching data,doing heaavy math  )
- save resourses :less cpu ,memory , or network usage.
- Makes program faster when the same inputs are used multiple times.

````python
#Example with out cache
def slow_square(n):
    print("calculating")
    return n*n
print(slow_square(5))#run once
print(slow_square(5))#recalculate everything 
````

````python
cache={}
def slow_square(n):
    if n in cache:
        return cache[n]
    print("Calculating....")
    result=n*n
    cache[n]=result
    return result
print(slow_square(5))
print(slow_square(5))## returns cached result 
print(slow_square(6))
print(slow_square(6))
````

**Above output shows** 
- if it is already calculated then it just return the value not recalculate everything

### Memonization 
*which we did in calcualting Fibb series , we did save result in cache so not to calculate everything for same input*
- It’s a technique where you remember (cache) the results of function calls.
- If the function is called again with the same input, you don’t recompute — you just return the stored result.
- This makes recursive algorithms (like Fibonacci) much faster because they avoid repeating the same work.


````python
table = {}
a = []

def fibM(n):
    if n == 0 or n == 1:
        return n
    else:
        if n not in table:              # check cache
            table[n] = fibM(n-1) + fibM(n-2)  # compute and store in cache
        return table[n]

n = int(input("enter the number of fib series: "))
for i in range(n):
    a.append(fibM(i))

print(*a)
print(table)
````

````python
# Create a decorator to cache results of a function (memoization).
from functools import wraps
import time#time.pref_cointer is best for to calculate execution time of function
def mytimer(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = func(*args, **kwargs)
        end = time.perf_counter()
        print(f"[TIMER] {func.__name__}({args[0]}) took {end - start:.6f}s")
        return result
    return wrapper

def memonize(func):
    cache={}
    @wraps(func)
    def wrapper(*args,**kwargs):
        key=(args,frozenset(kwargs.items()))#A forzenset  is just an immutable (unchangeable) version of a set.
        #forzen set is hasable unlike normal set so it can be used in dictonary
        if key in cache:
            return cache[key]
        result=func(*args,**kwargs)
        cache[key]=result
        return result
    return wrapper

@mytimer
@memonize
def fib(n):
    if(n<2):
        return n
    return fib(n-1)+fib(n-2)
print(fib(500))
print(fib(500))
````

# 🧠 Memoization with `@memoize` Decorator

## What Is Memoization?
Memoization is an optimization technique that **caches the results** of expensive function calls and returns the cached result when the same inputs occur again.

---

## The Decorator
```python
from functools import wraps

def memoize(func):
    cache = {}

    @wraps(func)
    def wrapper(*args, **kwargs):
        key = (args, frozenset(kwargs.items()))  # Create a unique, hashable key
        if key in cache:
            return cache[key]                    # Return cached result
        result = func(*args, **kwargs)           # Compute if not cached
        cache[key] = result                      # Store in cache
        return result

    return wrapper
```

---

## How the Key Works

| Component | Example | Purpose |
|-----------|---------|---------|
| `args` | `(10,)` | Positional arguments as a tuple |
| `frozenset(kwargs.items())` | `frozenset()` | Keyword arguments as an immutable set |
| `key` | `((10,), frozenset())` | Combined unique identifier for caching |

> 💡 `frozenset` is used because regular sets are **not hashable** and can't be used as dictionary keys.

---

## Call-by-Call Walkthrough: `square(10)`

### First Call
```
square(10)
  └─► wrapper(10)
        ├─ key = ((10,), frozenset())
        ├─ cache is EMPTY → miss
        ├─ compute square(10) → 100
        ├─ store: cache[((10,), frozenset())] = 100
        └─ return 100
```

### Second Call
```
square(10)
  └─► wrapper(10)
        ├─ key = ((10,), frozenset())
        ├─ key FOUND in cache ✅
        └─ return 100  ← instantly, no recomputation
```

---

## Cache State After First Call
```python
{
    ((10,), frozenset()): 100
}
```

---

## Applied to Fibonacci: `fib(n)`
```python
@memoize
def fib(n):
    if n < 2:
        return n
    return fib(n-1) + fib(n-2)

alist = [fib(i) for i in range(6)]
print(alist)  # [0, 1, 1, 2, 3, 5]
```

### Why Memoization Matters Here

Without memoization, `fib(5)` triggers an **exponential number of calls**:
```
fib(5) → fib(4) + fib(3)
           fib(4) → fib(3) + fib(2)   ← fib(3) computed AGAIN!
```

With memoization, each value is computed **exactly once** and reused.

| `n` | Without Memo | With Memo |
|-----|-------------|-----------|
| 5   | 15 calls    | 5 calls   |
| 10  | 177 calls   | 10 calls  |
| 30  | ~2.7M calls | 30 calls  |

---

## 🎒 The Notebook Analogy

| Action | Without Memo | With Memo |
|--------|-------------|-----------|
| First time solving | Compute the answer | Compute **and write it down** |
| Next time | Compute again 😓 | Just read the notebook 📖 |

---

## Key Takeaways

- `@memoize` wraps a function to **cache its outputs** by input
- The cache key combines `args` + `frozenset(kwargs)` for **hashability**
- Memoization turns **exponential** recursion (like Fibonacci) into **linear** time
- `@wraps(func)` preserves the original function's **name and docstring**
