```python
# Write a Python program to check if a given number is prime.
number=int(input("Enter the Number : "))
def prime():
    flag=False
    for i in range(2,(number//2)+1):
        if(number%i==0):
            flag=True
            break
    return flag
            
Check=prime()
if(Check==False):
    print(f"{number} is a prime number")
else:
    print(f"{number} is a composite number")
```

```python
# Implement a function to find the factorial of a number using recursion
def fact(n):
    if(n==1 or n==0):
        return 1
    else:
        return n*fact(n-1)

result=fact(5)
print(result)
```

```python
# Create a function that takes a list of numbers and returns only the even numbers.
def retEven(numbers):
    even=[]
    for number in numbers:
        if(number%2==0):
            even.append(number)
    return even
def reteven(*numbers):
    even=[]
    for i in numbers:
        if(i%2==0):
            even.append(i)
    return even
numbers=[1,2,3,4,5]
result=reteven(*numbers)
res=retEven(numbers)
print(result)
print(res)
```

```python
# Write a program to check if a string is a palindrome.
def checkPalindrome(text,length):
    for i in range(length//2):
        if(text[i]!=text[-(i+1)]):#text[0]=first ch of text and text[-1]=last ch of text
            return False
    return True        
text=input("Enter the string : ")
length=len(text)   
flag=checkPalindrome(text,length)
if(flag==True):
    print(f"The {text} is palindrome")
else:
    print(f"The {text} is not a palindrome")
```

```python
# Implement a class Circle with a method to calculate the area and circumference.
PI=3.14
class circle:
    def __init__(self,radius):
        self.radius=radius
    def area(self):
        Area=PI*(self.radius)**2
        return Area
    def circumference(self):
        p=PI*2*self.radius
        return p
c1=circle(2)
print(f"The Area of the circle is {c1.area():.2f}")
print(f"The perimeter of the circle is {c1.circumference():.2f}")
```

```python
# Create a class Rectangle with methods to calculate area and perimeter.
class Rectangle:
    def __init__(self,length,breadth):
        self.length=length
        self.breadth=breadth
    def __str__(self):
        return f"The Matrix of Rectangle is {self.length} Meter and {self.breadth} Meter Length and breadth respectively"
    def area(self):
        a=self.length*self.breadth
        return a
    def perimeter(self):
        p=2*(self.length+self.breadth)
        return p
R=Rectangle(4,3)
print(R)
print(f"Area of the rectangle is {R.area()} sq.meters")
print(f"The perimeter of the rectangle is {R.perimeter()} meters")
        
        
```

```python
# Implement a class Employee with attributes name and salary and a method to apply a 10% raise.
class Employee:
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary
    def __str__(self):
        return f"The Name of the employee is {self.name} and his salary is {self.salary}"
    def Raise(self):
        self.salary+=self.salary*0.1
E=Employee("Ayush",60000)
print(E)
E.Raise()
print("The employee get a raise of 10% : ")
print(E)
```
